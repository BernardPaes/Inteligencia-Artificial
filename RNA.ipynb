```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
```

```python
# Carregar a base de dados
data = pd.read_csv('breast-cancer.csv', delimiter=',')
print(data.isnull().sum())
```

Processamento da base de dados

```python
# Remove todas ocorrências de ?, substituindo por NaN
data.replace('?', np.nan, inplace = True)
X_comNAN = data.iloc[20, 0:10]
X_comNAN

# Separa o conjunto de dados de sua classificação
X_dados = data.iloc[:, 0:9].values
Y_classe = data.iloc[:, 9].values

# Trata os dados categóricos
label_node = LabelEncoder()
label_breast = LabelEncoder()
label_irradiat = LabelEncoder()

X_dados[:,0] = label_node.fit_transform(X_dados[:,0])
X_dados[:,1] = label_node.fit_transform(X_dados[:,1])
X_dados[:,2] = label_node.fit_transform(X_dados[:,2])
X_dados[:,3] = label_node.fit_transform(X_dados[:,3])
X_dados[:,4] = label_node.fit_transform(X_dados[:,4])
X_dados[:,6] = label_breast.fit_transform(X_dados[:,6])
X_dados[:,8] = label_irradiat.fit_transform(X_dados[:,8])

# Trata os dados não categóricos
onehotencoder_base = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [7])], remainder='passthrough')
X_dados = onehotencoder_base.fit_transform(X_dados)

X_dados
```

Divide o conjunto de dados em conjuntos de teste e treino

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X_dados, Y_classe, test_size = 0.2, random_state = 0)

X_treino.shape
```
